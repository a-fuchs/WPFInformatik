```micropython
%serialconnect


import os

def deleteFile( strFile ) :
    try:
        print( "remove file '{}'".format( strFile ) )
        os.remove( strFile)
    except Exception as _e:
        print( _e )
        #pass
    
def deleteDir( strDir ) :
    files = os.listdir(strDir)
    for f in files:
        deleteFile( strDir + "/" + f )

    deleteFile( strDir )

def lsDir( strDir, showContent = False ) :
    files = os.listdir(strDir)
    print( files )
    for f in files:
        print(f)
        if showContent :
            print( "Content >>---" )
            print(open(f).read())
            print( "-----------<<\n" )
    
```
