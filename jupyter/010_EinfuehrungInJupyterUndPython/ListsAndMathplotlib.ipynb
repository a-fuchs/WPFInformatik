```python
import matplotlib.pyplot as plt
import math

iN = 200

def isPrime( n ):
    for iNum in range( 2, int(math.sqrt(n)) + 1 ):
        if ( n % iNum == 0 ) : return False

    return True


aX = [x for x in range(iN)]
aY = [isPrime(x) for x in aX]

# or:
aY = list( map( isPrime, aX ) )

plt.figure(figsize=(20, 2))
plt.axis([0, iN, 0, 1.5])
plt.scatter(aX, aY, s=1, marker='o', color="black" )
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

iN = 200

class Primes:
    """Calculates primes"""
    def __init__(self):
        self.iCount = 0
        
    def reset(self):
        self.iCount = 0
    
    @staticmethod
    def isPrime( _iNum ) :
        if _iNum < 2 :
            return 0
        else :
            for iDivisor in range( 2, int(math.sqrt(_iNum)) + 1 ):
                if ( _iNum % iDivisor == 0 ) : return 0

            return _iNum
    
    def primesPercent(self, _iNum ) :
        if Primes.isPrime( _iNum ) :
            self.iCount+=1
    
        return self.iCount/_iNum
    
primes = Primes()


aX = [x + 1 for x in range(iN)]
aY = [primes.primesPercent(x) for x in aX]

plt.figure(figsize=(20, 2))
plt.axis([0, iN, 0, 1])
plt.scatter(aX, aY, s=1, marker='o', color="black" )
plt.show()
```
